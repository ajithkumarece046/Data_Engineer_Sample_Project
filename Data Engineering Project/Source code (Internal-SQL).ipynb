```python
import pandas as pd



import pyodbc




conn=pyodbc.connect('Driver={ODBC Driver 17 for SQL Server};'
                    'Server=BC-IN-HYD-DATA-\SQLEXPRESS01;'
                    'Database=Test_db;'
                    'Trusted_connection=yes;')

cursor=conn.cursor()



data=pd.read_csv('AdventureWorks Customer Lookup.csv',encoding='latin-1')
for index,row in data.iterrows():
    sql_insert=f"Insert into Customer_Table(CustomerKey,FirstName,LastName,BirthDate,MaritalStatus,Gender,EmailAddress,AnnualIncome,TotalChildren,EducationLevel,Occupation,HomeOwner) values(?,?,?,?,?,?,?,?,?,?,?,?)"

    cursor.execute(sql_insert, row['CustomerKey'], row['FirstName'], row['LastName'], row['BirthDate'], row['MaritalStatus'], row['Gender'], row['EmailAddress'], row['AnnualIncome'], row['TotalChildren'], row['EducationLevel'], row['Occupation'], row['HomeOwner'])    

data=pd.read_csv('AdventureWorks Product Subcategories Lookup.csv',encoding='latin-1')
for index,row in data.iterrows():
    sql_insert=f"Insert into Product_Subcategory_Table(ProductSubcategoryKey,SubcategoryName,ProductCategoryKey) values(?,?,?)"

    cursor.execute(sql_insert, row['ProductSubcategoryKey'], row['SubcategoryName'], row['ProductCategoryKey'])   
    
data=pd.read_csv('AdventureWorks Product Lookup.csv',encoding='latin-1')
for index,row in data.iterrows():
    sql_insert=f"Insert into Product_Table(ProductKey,ProductSubcategoryKey,ProductSKU,ProductName,ModelName,ProductDescription,ProductColor,ProductSize,ProductCost,ProductPrice) values(?,?,?,?,?,?,?,?,?,?)"

    cursor.execute(sql_insert, row['ProductKey'], row['ProductSubcategoryKey'], row['ProductSKU'], row['ProductName'], row['ModelName'], row['ProductDescription'], row['ProductColor'], row['ProductSize'],row['ProductCost'], row['ProductPrice'])    

    
data=pd.read_csv('AdventureWorks Returns Data.csv',encoding='latin-1')
for index,row in data.iterrows():
    sql_insert=f"Insert into Return_Product_Table(ReturnDate,TerritoryKey,ProductKey,ReturnQuantity) values(?,?,?,?)"

    cursor.execute(sql_insert, row['ReturnDate'], row['TerritoryKey'], row['ProductKey'], row['ReturnQuantity'])    

    
    
    
data=pd.read_csv('AdventureWorks Sales Data 2020.csv',encoding='latin-1')    
for index,row in data.iterrows():
    sql_insert=f"Insert into Sales_2020_Table(OrderDate,StockDate,OrderNumber,ProductKey,CustomerKey,TerritoryKey,OrderLineItem,OrderQuantity) values(?,?,?,?,?,?,?,?)"

    cursor.execute(sql_insert, row['OrderDate'], row['StockDate'], row['OrderNumber'], row['ProductKey'], row['CustomerKey'], row['TerritoryKey'], row['OrderLineItem'], row['OrderQuantity'])    

    
data=pd.read_csv('AdventureWorks Sales Data 2021.csv',encoding='latin-1')    
for index,row in data.iterrows():
    sql_insert=f"Insert into Sales_2021_Table(OrderDate,StockDate,OrderNumber,ProductKey,CustomerKey,TerritoryKey,OrderLineItem,OrderQuantity) values(?,?,?,?,?,?,?,?)"

    cursor.execute(sql_insert, row['OrderDate'], row['StockDate'], row['OrderNumber'], row['ProductKey'], row['CustomerKey'], row['TerritoryKey'], row['OrderLineItem'], row['OrderQuantity'])    


data=pd.read_csv('AdventureWorks Sales Data 2022.csv',encoding='latin-1')        
for index,row in data.iterrows():
    sql_insert=f"Insert into Sales_2022_Table(OrderDate,StockDate,OrderNumber,ProductKey,CustomerKey,TerritoryKey,OrderLineItem,OrderQuantity) values(?,?,?,?,?,?,?,?)"

    cursor.execute(sql_insert, row['OrderDate'], row['StockDate'], row['OrderNumber'], row['ProductKey'], row['CustomerKey'], row['TerritoryKey'], row['OrderLineItem'], row['OrderQuantity'])    

    
    
data=pd.read_csv('AdventureWorks Territory Lookup.csv',encoding='latin-1')        
for index,row in data.iterrows():
    sql_insert=f"Insert into Territory_Table(SalesTerritoryKey,Region,Country,Continent) values(?,?,?,?)"

    cursor.execute(sql_insert, row['SalesTerritoryKey'], row['Region'], row['Country'], row['Continent'])    

data = pd.read_csv('AdventureWorks Product Categories Lookup.csv', encoding='latin-1')

for index,row in data.iterrows():
    sql_insert=f"Insert into Product_Category_Table(ProductCategoryKey,CategoryName) values(?,?)"
    cursor.execute(sql_insert, row['ProductCategoryKey'], row['CategoryName'])
    

    
conn.commit()

cursor.close()

conn.close()
```
